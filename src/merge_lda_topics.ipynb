```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# load keywords 
df = pd.read_csv('news_data/keywords_all_states_improved.csv')
```

```python
# plot word clouds for each topic
topic_cols = [col for col in df.columns if col.startswith("Topic_")]
for i in range(0, len(topic_cols), 2):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    for j in range(2):
        if i + j < len(topic_cols):
            col = topic_cols[i + j]
            text = ' '.join(df[col].dropna().tolist())
            wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis', random_state=123).generate(text)
            axes[j].imshow(wordcloud, interpolation='bilinear')
            axes[j].axis('off')
            axes[j].set_title(col, fontsize=16)
    plt.tight_layout()
    plt.show()
```

We can observe some overlap between topics. 
We believe this is due to the nature of our data sources — we retrieved data from Media Cloud and Google News, which do not include all news published by local outlets.

> “Media Cloud collects most of its content through the RSS feeds of the media sources **we follow**. We only have data for a source from the time we started scraping its RSS feeds. Additionally, we complement the RSS data with sitemap data.” ([link here](https://www.mediacloud.org/documentation/faqs))

As stated directly on their website, Media Cloud only includes articles from the sources they follow, which may lead to underrepresentation of certain topics.

> “Google’s technology analyzes news across the web to determine the top stories that will show up for everyone with the same language settings in a given country, based primarily on what publishers are writing about. Once these stories are identified, algorithms then select which specific content to surface and link to for each story, based on factors such as the prominence and freshness of the piece of content, and authoritativeness of the source.” ([link here](https://www.google.com/search/howsearchworks/how-news-works/))

Similarly, Google’s algorithms select news based on specific criteria, which may also contribute to the underrepresentation of some sources.

We acknowledge that our sample is biased, as it does not fully represent all available news articles. However, we believe that the analysis we conducted still provides informative insights relevant to our research objectives.
